```python
def latex_float(f, n):
    kwrd = '{0:.' +str(n)+ 'g}'
    float_str = kwrd.format(f)
    if "e" in float_str:
        base, exponent = float_str.split("e")
        return r"{0} \times 10^{{{1}}}".format(base, int(exponent))
    else:
        return float_str
```
