The source function is given by
$$Q_e = \frac{dN}{dEdt}n_{MSP}(\boldsymbol{x})$$
$$ \equiv \mathcal{N} (E/E_{cut})^\alpha e^{-E/E_{cut}}n_{MSP}(\boldsymbol{x})$$

The power to electrons and positrons per millisecond pulsar is
$$ \eta_e P_{MSP} = \int_0^{\infty} dE E\frac{dN}{dE dt}$$
$$= \mathcal{N} \int_0^{\infty} dE E (E/E_{cut})^\alpha e^{-E/E_{cut}}$$
$$\implies \mathcal{N} = \frac{\eta_e P_{MSP}}{\int_0^{\infty} dE E (E/E_{cut})^\alpha e^{-E/E_{cut}}}$$
The luminosity of gamma rays coming from within a volume V is given by
$$L_{\gamma, V} = \eta_\gamma P_{MSP}\int_V dV n_{MSP}$$
allowing us to solve for $P_{MSP}$:
$$P_{MSP} = \frac{L_{\gamma, V}}{\eta_\gamma\int_V dV n_{MSP}} = \frac{L_{\gamma, V}}{\eta_\gamma V \langle n_{MSP}\rangle_V}$$

Substituting this into the expression for $\mathcal{N}$ and the expression for $\mathcal{N}$ into the expression for $Q_e$ gives

$$Q_e = \frac{\eta_e}{\eta_\gamma} \frac{L_{\gamma, V}}{V \int_0^{\infty} dE E (E/E_{cut})^\alpha e^{-E/E_{cut}}} \frac{n_{MSP}(\boldsymbol{x})}{\langle n_{MSP}\rangle_V}(E/E_{cut})^\alpha e^{-E/E_{cut}}$$

```python
GeVpererg = 624

alpha = 1.7
L_gamma_erg = 2e37 #erg/s
R = 5 #kpc
ratio = 1 #energy into e over energy into gamma
cm_per_kpc = 3.09e21

V = (4*np.pi/3)*R**3
L_gamma = L_gamma_erg*GeVpererg


#rate of injection of electrons and positrons per energy per MSP 
def dN_dEdt(E, norm=1):
    return norm*((E/Ecut)**alpha)*np.exp(-E/Ecut)

#power to electrons and positrons per MSP
def EdN_dEdt(E):
    return E*dN_dEdt(E)

#power per MSP divided by norm (the integral in the denominator of markdown cell above)
Pmsp_norm = quad(EdN_dEdt, 0, Emax)[0]
```

```python
# source averaged over V as a function of energy
def Q_V(E):
    return dN_dEdt(E, norm=ratio*L_gamma/(V*Pmsp_norm*cm_per_kpc**3))
```

When energy loss dominates the dynamics over diffusion, the phase space density can be written as
$$f_e = \frac{1}{b(E)}\int_E^{E_{max}} dE' Q_e(r, E')$$
The synchrotron flux is then
$$\frac{d^2 S}{d\nu d\Omega} = \frac{1}{4\pi}\int dl \int dE f_e(E, r) \langle \frac{dP}{d\nu}\rangle (r, E, \nu)$$
$$= \frac{1}{4\pi}\int dl \int_{E_{min}}^{E_{max}} dE \frac{\langle\frac{dP}{d\nu}\rangle (r, E, \nu)}{b(E)}\int_E^{E_{max}}dE' Q_e(r, E') $$
$\langle dP/d\nu\rangle$ is in terms of $x=x_0/E^2$ so I will substitute for $x$.
