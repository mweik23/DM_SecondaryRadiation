```python
import os
import numpy as np
import time
from scipy.stats import percentileofscore
```

```python
base_path = os.getcwd().split('scripts')[0]
dchi_path = base_path + 'dchi/'
dchi_list = os.listdir(dchi_path)
#take only the new files
dchi_list = [dchi_element for dchi_element in dchi_list if '_numsamps_' in dchi_element]
```

```python
num_mx = 20
mx_min = 6
mx_max = 500
num_sigmav = 80
log_sigma_min = -28
log_sigma_max = -25
mx_arr = np.round(np.logspace(np.log10(mx_min), np.log10(mx_max), num_mx), 1)
sigmav_arr = np.logspace(log_sigma_min, log_sigma_max, num_sigmav)
D0 = 3e28
D0_str = str(D0)
dchi_sb_file_names = [[] for i in range(num_mx)]
dchi_b_file_names = [[] for i in range(num_mx)]
```

```python
for file in dchi_list:
    
    #check if the diffusion parameter is correct
    D0_file = file.split('_D0_')[1].split('_')[0]
    #print('value of D0_file ' + D0_file)
    if D0_file==D0_str:
        
        #check if it is sb or b
        prefix = file.split('_starting_sample')[0]
        if prefix == 'dchi_sb':
            sb_file = True
        elif prefix == 'dchi_b':
            sb_file = False
        else:
            print('dchi prefix is not recognized')
    
        #check what the mass is
        mx_file = np.float(file.split('_mx_')[1].split('_')[0])
        mx_ind = np.where(mx_arr==mx_file)[0]
        
        if len(mx_ind)>0:
            if sb_file:
                dchi_sb_file_names[mx_ind[0]].append(file)
            else:
                dchi_b_file_names[mx_ind[0]].append(file)
    
    
```

```python
dchi_sb_llt = [[tuple(dchi_sb_ll[i][j][:, l] for j in range(len(dchi_sb_file_names[i]))) \
                 for l in range(num_sigmav)] for i in range(len(dchi_sb_file_names))]

dchi_b_llt = [[tuple(dchi_b_ll[i][j][:, l] for j in range(len(dchi_b_file_names[i]))) \
                 for l in range(num_sigmav)] for i in range(len(dchi_b_file_names))]
dchi_sb_comb = np.array([[np.concatenate(dchi_sb_llt[i][l]) for l in range(num_sigmav)] for i in range(num_mx)])
dchi_b_comb = np.array([[np.concatenate(dchi_b_llt[i][l]) for l in range(num_sigmav)] for i in range(num_mx)])
```

```python
print(dchi_sb_comb.shape)
print(dchi_b_comb.shape)
dchi_comb_path = base_path + 'dchi_combined/'
dchi_comb_suffix = 'samples_0_29999_D0_3e28_mx_logspace_'+str(mx_min)+'_'+str(mx_max)+'_'+str(num_mx)+\
'_round1dec_sigmav_logspace_'+str(log_sigma_min)+'_'+str(log_sigma_max)+'_'+str(num_sigmav)
np.save(dchi_comb_path + 'dchi_sb_' + dchi_comb_suffix + '.npy', dchi_sb_comb)
np.save(dchi_comb_path + 'dchi_b_' + dchi_comb_suffix + '.npy', dchi_b_comb)
```
